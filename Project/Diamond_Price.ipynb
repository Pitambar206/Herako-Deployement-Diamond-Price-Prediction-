# Data Preparation:
* Train Test Split
* Encoding for Categorical Columns:
     - Ordinal : LabelEncoding or OrdinalEncoding
     - Nominal : OneHotEncoding or get_dummies

* Encoding for Numerical Columns
   - Standardization (z-transformation)
    
**We will be following below mentioned steps:**

* a. Identify the Target Variable and Splitting the Data into train and test
* b. Separating Categorical and Numerical Columns
* c. Rescaling Numerical Columns (Standardization or z-transformation)
* d. Applying OneHotEncoding on Categorical Columns
* e. Applying Label Encoding on Categorical Columns
* f. Concatinating the Encoded Categorical Features and Scaled Numerical Features

```python
X=df.drop(['price'],axis=1)
X
```

```python
y=df['price']
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.7)
```

```python
X_train.head()
```

```python
X_train_cat_le=pd.DataFrame(index=X_train.index)
X_train_cat_le.head()
```

```python
X_train['cut'].unique()
```

```python
cut_encode={'Fair':1,'Good':2,'Very Good':3,'Ideal':4,'Premium':5}
X_train_cat_le['cut']=X_train['cut'].apply(lambda k :cut_encode[k])
X_train_cat_le.head()
```

```python
X_train['clarity'].unique()
```

```python
clarity_encode={'I1':1, 'SI2':2, 'SI1':3, 'VS2':4, 'VS1':5, 'VVS2':6, 'VVS1':7, 'IF':8}
X_train_cat_le['clarity']=X_train['clarity'].apply(lambda y :clarity_encode[y])
X_train_cat_le.head()
```

```python
X_train['color'].unique()
```

```python
color_encode={'J':1,'I':2,'H':3,'G':4,'F':5,'E':6,'D':7}
X_train_cat_le['color']=X_train['color'].apply(lambda y :color_encode[y])
X_train_cat_le.head()
```

```python
X_train_num=X_train.select_dtypes(include=['int64','float64'])
X_train_num.head()
```

```python
# scaling the numerical features
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

# column names are (annoyingly) lost after Scaling
# (i.e. the dataframe is converted to a numpy ndarray)

X_train_num_rescaled = pd.DataFrame(scaler.fit_transform(X_train_num), 
                                    columns = X_train_num.columns, 
                                    index = X_train_num.index)

X_train_num_rescaled.head()
```

```python
sns.displot(X_train_num_rescaled['table'])
```

```python
X_train_trf=pd.concat([X_train_num_rescaled,X_train_cat_le],axis=1)
X_train_trf.head()
```

```python
X_test.head()
```

```python
X_test_num=X_test.select_dtypes(include=['int','float64'])
X_test_num.head()
```

```python
X_test_cat_le=pd.DataFrame(index=X_test.index)
X_test_cat_le.head()
```

```python
X_test_cat_le['cut']=X_test['cut'].apply(lambda k : cut_encode[k])
X_test_cat_le.head()
```

```python
X_test_cat_le['clarity']=X_test['clarity'].apply(lambda y : clarity_encode[y])
X_test_cat_le['color']=X_test['color'].apply(lambda y : color_encode[y])
```

```python
X_test_cat_le.head()
```

```python
X_test_num_rescaled = pd.DataFrame(scaler.transform(X_test_num), 
                                   columns = X_test_num.columns, 
                                   index = X_test_num.index)

X_test_num_rescaled.head()

```

```python
X_test_trf=pd.concat([X_test_num_rescaled,X_test_cat_le],axis=1)
X_test_trf.head()
```

```python
from sklearn.linear_model import LinearRegression
regressor= LinearRegression()
regressor.fit(X_train_trf,y_train)
```

```python
y_test_pred=regressor.predict(X_test_trf)
y_test_pred
```

```python
temp_df = pd.DataFrame({'Actual': y_test, 'Predicted': y_test_pred})

temp_df.head()
```

```python
from sklearn.neighbors import KNeighborsRegressor
K_n= KNeighborsRegressor()
K_n.fit(X_train_trf,y_train)
```

```python
y_pred=K_n.predict(X_test_trf)
```

```python
temp_df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})

temp_df.head()
```
